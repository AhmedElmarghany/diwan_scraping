```python
import time
import requests
from bs4 import BeautifulSoup
import re
import os
```

```python
odd_lines_quotes = []
for page in range(1, 69):
    url = f'https://www.aldiwan.net/quotes.html?page={page}'
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html')
    cards = soup.select("div.quote")
    for card in range(len(cards)):
        # Quote
        quote = soup.select('div.quote > p.quote-info')[card]
        quote = quote.get_text(separator = '\n', strip = True)
        quote = list(quote.split('\n'))
        if len(quote) % 2 == 1:
            odd_lines_quotes.append(quote)
        # Topic
        topic = soup.select('div.quote > div.d-inline-block > a:nth-of-type(1)')[card].text.strip()
        topics.append(topic)
        # Author
        author = soup.select('div.quote > div.d-inline-block > a:nth-of-type(2)')[card].text.strip()
        authors.append(author)
        # writing in file
        with open('quotes.txt', 'a', encoding="utf-8") as f:
            for line in quote:
                f.write("%s\n" % line)
            f.write(topic + "\n")
            f.write(author + "-" + "\n")
            f.write("="*20 + "\n")
            f.close()
    if page % 10 == 0:
        print(f"__ {page} pages done __")
print("__Scraping done Successfully__")
```
